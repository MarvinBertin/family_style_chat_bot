```python
users = ['Brian', 'David', 'Eugene', 'Jon', 'Marvin', 'Bob', 'Anne', 'Sarah', 'Leah', 'Samatha', 'Michelle']
cuisines = ['Lime Tree Southeast Asian Kitchen', 'La Urbana', 'Jayakarta Supreme', 'Chapeau!', 
            'Shanghai Dumplings', 'Saigon Cafe', 'Sanraku', 'Little Delhi', 'Lers Ros Thai', 
            "Tommaso's Ristorante Italiano", 'Pizza']
#cuisines = ['China', 'France', 'India', 'Indonesia', 'Italy', 'Japan', 
#            'Mexico', 'Pizza', 'Singapore', 'Thailand', 'Vietnam']

possible_cuisine_or_dish_rating = np.arange(0.5, 5.5, 0.5)


def generate_ratings(length, possible_scores=np.arange(0.5, 5.5, 0.5)):
    temp = np.random.choice(possible_scores, length)
    return temp * np.random.randint(2, size=length)

def dish_scorer(number_of_people, number_of_dishes):
    return generate_ratings((number_of_people, number_of_dishes))
```

```python
user_cuisine_rating = pd.DataFrame({'user_id': users})

for cuisine in cuisines:
    user_cuisine_rating[cuisine] = None

user_cuisine_rating.ix[:, 1:] = generate_ratings(user_cuisine_rating.shape - np.array([0, 1]))
user_cuisine_rating = pd.melt(user_cuisine_rating, id_vars='user_id', var_name='item_id', value_name='rating')
user_cuisine_rating = user_cuisine_rating.ix[user_cuisine_rating['rating'] != 0, :]
user_cuisine_rating.head()
```

```python
cuisine_user_dish_tables = {}
for cuisine in cuisine_dish_list:
    temp = pd.DataFrame({'user_id':users})
    for dish in cuisine_dish_list[cuisine]:
        temp[dish] = None

    temp.ix[:, 1:] = generate_ratings(temp.shape - np.array([0, 1]))
    temp = pd.melt(temp, id_vars='user_id', var_name='item_id', value_name='rating')        
    temp = temp.ix[temp['rating'] != 0, :]
    cuisine_user_dish_tables[cuisine] = temp
```
