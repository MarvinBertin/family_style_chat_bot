```python
import pandas as pd
import numpy as np
np.random.seed(1234)
```

```python
users = ['Brian', 'David', 'Eugene', 'Jon', 'Marvin', 'Bob', 'Anne', 'Sarah', 'Leah', 'Samatha', 'Michelle']
cuisines = ['China', 'France', 'India', 'Indonesia', 'Italy', 'Japan', 
            'Mexico', 'Pizza', 'Singapore', 'Thailand', 'Vietnam']
possible_cuisine_or_dish_rating = np.arange(0.5, 5.5, 0.5)


def generate_ratings(length, possible_scores=np.arange(0.5, 5.5, 0.5)):
    return np.random.choice(possible_scores, length)

def dish_scorer(number_of_people, number_of_dishes):
    return generate_ratings((number_of_people, number_of_dishes))
```

```python
cuisine_dish_list = {"Singapore": ['Sesame Oil Chicken Wings', 'Singapore Satay', 'Singapore Chili Crabs',
 'Singapore Noodle Curry Shrimp', 'Singapore Sling Cocktail', 'Singapore Laksa'],
 
 'Mexico': ['Great Chicken Taquitos', 'Double Chili cheese Dip', 'Grilled Fish Tacos', 
'Chili Rellenos', 'Albondigas', 'Vegetarian Tortilla Soup'],
 
 'Indonesia': ['Indonesian Banana Fritters', 'Indonesia Satay', 'Indonesian Spiced Rice', 
               'Indonesian Fried Noodles', 'Indonesian Beef Rendang', 'Oxtail soup'],
                     
 'France': ['Peach and Goat Cheese Tartine', 'Creme brulee', 'Crepe Suzette', 'Ratatouille',
            'French Onion Soup', 'French Beef Stew'],
                     
  'China': ['Chinese Tea Leaf Eggs', 'Seseame Oil Chicken Wings', 'Chinese Mabo Tofu',
             "General Tsao's Chicken", 'Chinese Sizzling Rice Soup', 'Chinese Corn Soup'],
                     
 'Vietnam': ['Vietnamese Fresh Spring Rolls', 'Vietnamese Salad Rolls', 'Vietnamese Chicken Salad', 
             'Vietnamese Aromatic Lamb Chops', 'Vietnamese Beef Pho', 'Vietnamese Style Vegetarian Curry Soup'],
                     
 'Japan': ['Japanese-Style Grilled Mushrooms', 'Tofu Hiyayakko', 'Crispy Shrimp Tempura', 
           'Miso Salmon', 'Dashi Stock', 'Beef Sukiyaki'], 

                     
 'India': ['Crispy Vegetables Pakoras', 'Green Chutney', 'Indian Pizza', 'Curried Chickpeas', 
           'Indian Chai', 'Mango-Mint Lassi'],

 'Thailand': ['Fried Chicken Wings', 'Fried Tofu with Peanut Dipping Sauce', 'Barbeque Pork on Rice',
              'Beef Stew Recipe', 'Thai Ginger Soup', 'Thai Hout and Sour Soup'],

 'Italy': ['Caprese Appetizer', 'Prosciutto Cups', 'Shrimp Scampi with Pasta', 'Pork Chops Italiano',
        'Italian Halibut Chwoder', 'Creamy Italian White Bean Soup'],
                     
 'Pizza': ['Berry, Arugula and Prosciutto Pizza', 'Macaroni And Cheese Pizza', 
           'Butternut Squash and Crispy Sage Pizza', 'Buffalo Chicken Pizza Sticks',
          'Cantaloupe And Sweet Ricotta Pizza', 'Hummus And Grilled Zucchini Pizzas']
            
}
```

```python
user_cuisine_rating = pd.DataFrame({'user_id': users})

for cuisine in cuisines:
    user_cuisine_rating[cuisine] = None

user_cuisine_rating.ix[:, 1:] = generate_ratings(user_cuisine_rating.shape - np.array([0, 1]))
user_cuisine_rating
```

```python
user_cuisine_rating = pd.melt(user_cuisine_rating, id_vars='user_id', var_name='item_id', value_name='rating')
```

```python
cuisine_user_dish_tables = {}
for cuisine in cuisine_dish_list:
    temp = pd.DataFrame({'user_id':users})
    for dish in cuisine_dish_list[cuisine]:
        temp[dish] = None

    temp.ix[:, 1:] = generate_ratings(temp.shape - np.array([0, 1]))    
    temp = pd.melt(temp, id_vars='user_id', var_name='item_id', value_name='rating')        
    cuisine_user_dish_tables[cuisine] = temp
```
